## Removing irrelevant columns

```python
# if inplace is false we don't need to print the dataset separately it will be printed automatically when drop function is completed
dataset.drop(['Name','SibSp','Ticket'], axis=1,inplace=True)
dataset
```

## Handling missing data

```python
# information about the dataset
dataset.info()
```

### method 1 - remove missing values by a created method

```python
# remove columns which have more than 30% of missing values
colNum = 0
# no.of rows
TotalObjects = dataset.shape[0]
print("Column\t\t\t\t\t Null Values %")
for row in dataset:
    nullCount = dataset[row].isnull().sum()
    nullPercentage = nullCount * 100 / TotalObjects
    if nullCount > 0 and nullPercentage > 30:
        colNum = colNum + 1
        dataset.drop(row,axis=1, inplace=True)
        print(str(row)+"\t\t\t\t\t " + str(nullPercentage))
print("A total of " + str(colNum) + " deleted!")
```

### method 2 - replace null values

```python
# replace the NaN with string Unknown
dataset['Embarked'].fillna('Unknown', inplace=True)

# replace the NaN with 0
dataset['Age'].fillna(0, inplace=True)
dataset['Ticket Booking Period'].fillna(0, inplace=True)
```

### method 3 - replace null values with mean

```python
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
```

### confirming no null values

```python
dataset.isnull().sum()
```

## Formatting data

```python
# Ticket booking period column of the data frame contains information about ticket reservation
pd.unique(dataset['Ticket Booking Period'].values)
```

```python
# changing to numeric
def calTicketBookingPeriod(week):
    if week == '< 1 week':
        return 0.5
    elif week == '4+ weeks':
        return 4
    else:
        wk = str(week)
        return wk.rstrip('weeks')
```

```python
# applying function to column
dataset['Ticket Booking Period'] = dataset['Ticket Booking Period'].apply(calTicketBookingPeriod)
dataset['Ticket Booking Period']
```

## Dealing with categorical data

```python
def getMaleNumericValue(str):
    if str == "male":
        return 1
    else:
        return 2

dataset['Gender'] = dataset["Gender"].apply(getMaleNumericValue)
dataset['Gender'].head()
```

```python
def getEmbarkedNumericValue(str): 
    if str == "S": 
        return 1 
    elif str == 'Q': 
        return 2 
    else: 
        return 3 

dataset["Embarked"] = dataset["Embarked"].apply(getEmbarkedNumericValue) 
dataset['Embarked'].head()
```

## Cleaned dataset

```python
dataset.to_csv("Titanic_cleaned.csv", index = False)
```

### Male Female Pie chart

```python
# male count
males = (dataset['Gender'] == 1).sum()  
# female count
females = (dataset['Gender'] == 2).sum() 

countArray = [males, females] 
print(countArray)

# plotting the pie chart
plt.pie(
    countArray, # count array
    labels = ['Male', 'Female'],  # Labels for pie chart  
    colors = ['green', 'yellow'], # colours for pie chart 
    explode = (0.10, 0), # gap between the pies 
    startangle = 90)  # what start angle should be given 
plt.axis('equal')  
plt.show() 
 
```
