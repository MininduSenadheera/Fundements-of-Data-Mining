# Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

# Importing the dataset

```python
df = pd.read_csv("income.csv")
df
```

# Scatter plot for the dataset

```python
# to scatter plot we need numerical values so the name is ignored
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

# Clustering data

```python
# 3 means 3 clusters
km = KMeans(3)
# fitting the data to above model to predict
y_predicted = km.fit_predict(df[['Age','Income($)']])
# 3 clusters have been created clusters of 0s 2s and 1s
y_predicted
# adding cluster to the data frame
df['cluster'] = y_predicted
df
```

# Plotting the scatter plot according to clusters

```python
# c is set as the cluster column so colours varies with the cluster
# cmap is used to add colours for different cluster
plt.scatter(df['Age'],df['Income($)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Income($)')
```

# Above clustering has some errors as you can see. Resolving the errors and re clustering

## Scaling Income to 0 - 1 values

```python
scaler = MinMaxScaler()

# choose income as the scaling column
scaler.fit(df[['Income($)']])
# scaling the income by dividing each income by the maximum income
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

## Scaling Age to 0 - 1 values

```python
# choose age as the scaling column
scaler.fit(df[['Age']])
# scaling the Age by dividing each Age by the maximum Age
df['Age'] = scaler.transform(df[['Age']])
df
```

## Getting the prediction after scaling

```python
# 3 means 3 clusters
km = KMeans(3)
# fitting the data to above model to predict
y_predicted = km.fit_predict(df[['Age','Income($)']])
# 3 clusters have been created clusters of 0s 2s and 1s
y_predicted
# adding cluster to the data frame
df['cluster'] = y_predicted
df
```

## Plotting again after re scaling

```python
# c is set as the cluster column so colours varies with the cluster
# cmap is used to add colours for different cluster
plt.scatter(df['Age'],df['Income($)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Income($)')

# now clusters has been divided properly
```

## Getting the cluster centroids

```python
km.cluster_centers_
```

# Scatter plot with the centroids

```python
# c is set as the cluster column so colours varies with the cluster
# cmap is used to add colours for different cluster
plt.scatter(df['Age'],df['Income($)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Income($)')
# x axis --> km.cluster_centers_[:,0]
# y axis --> km.cluster_centers_[:,1]
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroids')
```

# Implementation of KMeans Algorithm

```python
import pandas as pd 
import numpy as np  
from matplotlib import pyplot as plt  
%matplotlib inline 
```

```python
sse = []
# iterate from 1 to 9
k_rng = range(1,10)

for k in k_rng:
    # in each iteration, create a new model with cluster = k
    km = KMeans(n_clusters = k)
    # create a new frame excluding the mane col rather than passing the column names every
    # km.fit(<dataframe>) 
    km.fit(df[['Age','Income($)']])
    # inertia will give the sse
    sse.append(km.inertia_)

plt.xlabel('K')  
plt.ylabel('Sum of squared error') 
plt.plot(k_rng,sse) 

# the k value of the bending point is the minimum value of clusters that can be made
```
